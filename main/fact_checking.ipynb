# Fact-checking and preprocessing

In this notebook, I preprocessed the tweet json obejcts to select tweet level, user_level information related columns that may be useful in distinguishing fake news and true news. Then I fact checked each tweet and created a new column to store the result. I saved the output to google cloud storage for future use.This notebook serves as preprocessing step for subsequent analysis.

The tweet level features selected are: 'id_str','timestamp_ms','created_at' (when the tweet was created),'favorite_count','retweet_count','coordinates','geo' (the location the tweet was sent, use this instead of coordinates or place),'place','possibly_sensitive' (does it contain sensitive information, things like child abuse)

The user level features selected are: 'id_str','created_at','screen_name','name','verified', 'followers_count','friends_count','favourites_count','statuses_count','time_zone'

In addition, I selected raw texts. If the tweet is a retweet, I selected the original tweet id and the user id for the original tweet which will be useful in building a retweet network for network analysis.
  


Fake news is defined as tweets that contain an url that points to a fake news source and true news is defined as tweets that contain an url that points to a credible news website.For fact checking,I subsetted tweets that contain links to external news websites. I then classfied the individual links shared on Twitter as fake or true based on a combined list of domains (see domains.csv) that were producing mostly misformation compiled by Zimdars (2016)--used in other misinformation studies--and a most recent list provided by ​Guess, Nagler, and Tucker ​(2018). If the tweet is true, it has a score 1. If the tweet is false, it has a score -1.If the tweet does not contain an url, it is classified as None and if it contains an url not in true or fake news domain, it has a score 0.

I will define the fact-checking function below. Largely, the idea is to compile a list of known fake news websites and check if the url in a tweet contains points to the website. And the list of fake and true news websites are stored in domains.csv

```python
#read in true news and fake news domains
domains = sc.textFile('gs://good-bucket/nicole/domains.csv')
domains = domains.collect()
fake_news_domain = [i.split(',')[0] for i in domains[1::]]
true_news_domain = [i.split(',')[1] for i in domains[1::]][:38] 
```

Below I will write some user define functions to extract/create useful columns from tweet json object.


```python
#extract text through user defined function
def text_feature(truncated, text, extended_tweet):
    
    if truncated:
        return extended_tweet['full_text']
    else:
        return text
text_udf = udf(lambda x,y,z: text_feature(x,y,z), StringType())
```

```python
#to get user features, need to return a struct object, to do so, need to manually input schema

schema = StructType([
    StructField("user_id_str", StringType(), True),
    StructField("user_crearted_at", StringType(), True),
    StructField("user_screen_name", StringType(), True),
    StructField("user_name", StringType(), True),
    StructField("user_verified", BooleanType(), True),
    StructField("user_followers_count", IntegerType(), True),
    StructField("user_friends_count", IntegerType(), True),
    StructField("user_favourites_count", IntegerType(), True),
    StructField("user_statuses_count", IntegerType(), True),
    StructField("user_timezone", StringType(), True),
])


def get_user_features(user):
    
    result = []
    for i in user_features:
        try:
            result.append(user[i])
        except:
            result.append(None)
    return result

user_feature_udf = udf(get_user_features, schema)    
```

```python
#fact checking function

def fact_check(url, fake_domains=fake_news_domain , true_domains= true_news_domain):

    '''
    input a string of url, a list of fake domains, and a list of true domains
    output a score: -1: fake news domain; 0: neutral domain; 1: credible domain
    '''
    if url == None:
        return None
    
    url = url.lower()
    score = None

    # exclude twitter internal reference links
    if 'twitter.com/' in url:
        score = None

    else:

        # check if it is fake
        for f_domain in fake_domains:
            if f_domain in url:
                score = -1
                break

        # if it is not fake, check if it is true
        if score != -1:
            for t_domain in true_domains:
                if t_domain in url:
                    score = 1
                    break

        # if it is not fake nor true: neutral
        if score != 1 and score != -1:
            score = 0

    return score
```

```python
#create fact check column

new_schema = StructType([
    StructField("url", StringType(), True),
    StructField("fact_check", IntegerType(), True),
])


def get_url(truncated,extended_tweet,entities):
    
    tweets = {}
    tweets['truncated'] = truncated
    tweets['extended_tweet'] = extended_tweet
    tweets['entities'] = entities
    
    try:
        if tweets['truncated'] and tweets['extended_tweet']['entities']['urls'] != []:
            url = tweets['extended_tweet']['entities']['urls'][0]['expanded_url']
        elif 'unwound' in tweets['entities']['urls'][0]:
            url = tweets['entities']['urls'][0]['unwound']['url']
        else:
            url = tweets['entities']['urls'][0]['expanded_url']
    except:
        url = None

    return url, fact_check(url)

url_udf = udf(get_url, new_schema)    
```

```python
#user define function to get retweet id

def retweeting_id(retweeted_status):
    try:
        retweeted_status['id_str']
        return retweeted_status['id_str']
    except:
        return None
retweeting_id_udf = udf(lambda x: retweeting_id(x), StringType())
```

```python
#user define function to get retweet user id

def retweeting_user_id(retweeted_status):
    try:
        retweeted_status['user']['id_str']
        return retweeted_status['user']['id_str']
    except:
        return None
retweeting_user_id_udf = udf(lambda x: retweeting_user_id(x), StringType())
```

Now I will use all above functions to create/select useful columns and write the result to new json files in preparation for subsequent analyses.

```python
#use pyspark sql query to create/select columns
new_dataframe = data_from_file.withColumn('text',text_udf(data_from_file.truncated,data_from_file.text,data_from_file.extended_tweet)).withColumn('retweeting_id',retweeting_id_udf(data_from_file.retweeted_status)).withColumn('retweeting_user_id',retweeting_user_id_udf(data_from_file.retweeted_status))
newer_dataframe= new_dataframe.select('*',url_udf('truncated','extended_tweet','entities').alias("foobar"),user_feature_udf('user').alias('good')).select(*tweet_features,'text', "foobar.*","good.*",'retweeting_id','retweeting_user_id')
```

Some Explanatory data analysis to know the data we stored

```python
def fake_true_count(folder):
    
    '''Input a folder, return the number of fake/ture/neutral tweets passed for each file in that folder'''
    
    number_of_true_news = [0]*21
    number_of_fake_news = [0]*21
    number_of_neutral_news = [0]*21

    for i in range(21):
        data_from_file = spark.read.json(folder+files[i])
        number_of_true_news[i] = data_from_file.filter('fact_check==-1').count()
        number_of_fake_news[i] = data_from_file.filter('fact_check==1').count()
        number_of_neutral_news[i] = data_from_file.filter('fact_check==0').count()
    
    return number_of_true_news, number_of_fake_news, number_of_neutral_news
```

```python
number_of_true_news_hillary, number_of_fake_news_hillary, number_of_neutral_news_hillary = fake_true_count(hillary_folder)

number_of_true_news_trump, number_of_fake_news_trump, number_of_neutral_news_trump = fake_true_count(trump_folder)
```

```python
#plot of number of true/fake news parsed mentioning hillary or trump over the 20 day period

datelist = [i[:10] for i in files]
converted_dates = list(map(datetime.datetime.strptime, datelist, len(datelist)*['%Y-%m-%d']))
x_axis = converted_dates
formatter = dates.DateFormatter('%Y-%m-%d')

pyplot.figure(figsize=(20,15))
pyplot.subplot(2, 1, 1)
y_axis = number_of_true_news_hillary
pyplot.plot( x_axis, y_axis, '-', label = 'true' )

y_axis = number_of_fake_news_hillary
pyplot.plot( x_axis, y_axis, '-',label = 'fake' )

ax = pyplot.gcf().axes[0] 
ax.xaxis.set_major_formatter(formatter)
pyplot.gcf().autofmt_xdate(rotation=25)
pyplot.legend(loc='best',fontsize=15)
pyplot.title('Number of true and fake tweets mentioning hillary over the period',fontsize=15)

pyplot.subplot(2, 1, 2)

y_axis = number_of_true_news_trump
pyplot.plot( x_axis, y_axis, '-', label = 'true' )

y_axis = number_of_fake_news_trump
pyplot.plot( x_axis, y_axis, '-',label = 'fake' )

ax = pyplot.gcf().axes[0] 
ax.xaxis.set_major_formatter(formatter)
pyplot.gcf().autofmt_xdate(rotation=25)
pyplot.legend(loc='best',fontsize=15)
pyplot.title('Number of true and fake news mentioning trump over the period',fontsize=15)

pyplot.show()
```
