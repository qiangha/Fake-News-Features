In this notebook, I checked if some tweet and user level characteristics can be used as features to distinguish true or fake news. For tweet level, the features I considered were geo (the place of the tweet), created_at(the time of the tweet). For user_level, the features I considered were user verification, account creation time, favourite_count,friends_count, follower_count, statues_count.

For categorical variables like geo, created_at, I calculated the ratio of fake news to true news per category per date and plotted the result.

For continuous variables, like user_favourite_count, I built several logistic regression models with or without a given feature and compare model performance to select the features that are most relevant in classifying the news on Twitter.

## Tweet level features 
I will explore two tweet level features, locaton of the tweet (geo) and creation time for the tweet (created_at) to see if they could be used to classify news on twitter network.

### Tweet Location (where the tweet was sent)





```python
def get_city(place):
    try:
        place['full_name']
        return place['full_name']
    except:
        return None
    
get_city_udf = udf(lambda x: get_city(x), StringType())
```

```python
def city_useful(folder):
    '''Return the median of fake to true news ratio for all cities for the 21 days period.
    input is a directory, output is a dictionaryï¼Œkeys are cities, values are lists
    '''    
    
    ratio_record = {}

    for i in range(21):
        
        place_dict = {}
        data_from_file = spark.read.json(folder+files[i])
        new_dataframe = data_from_file.withColumn('place',get_city_udf(data_from_file.place)).filter('fact_check==1 or fact_check == -1').filter('place is not NULL').groupby('place','fact_check')
        records = new_dataframe.agg(func.count(func.lit(1)).alias("Num Of Records"))

        true_record = records.filter('fact_check==1').collect()
        false_record = records.filter('fact_check==-1').collect()

        for j in true_record:
            temp_place = j[0]
            if temp_place not in place_dict:
                place_dict[temp_place] = j[2]
        
        for k in false_record:
            temp_place = k[0]
            if temp_place in place_dict:
                if temp_place in ratio_record:
                    ratio_record[temp_place].append(k[2]/place_dict[temp_place])                
                else:
                    ratio_record[temp_place] = [k[2]/place_dict[temp_place]]
            
    return ratio_record
```

```python
ratio_record_hillary = city_useful(hillary_folder)
ratio_record_trump = city_useful(trump_folder)
```

```python
#plotted the results for 30 randomly selected cities for tweets mentioning hillary
figure(figsize=(20, 15))
pyplot.ylim(0, 5)
pyplot.boxplot(list(ratio_record_hillary.values())[:30],labels=list(ratio_record_hillary.keys())[:30], notch = False)
pyplot.title('Boxplot for ratio of fake news against true news for different cities (hillary)',fontsize = 20)
pyplot.ylabel('Ratio of fake to true news',fontsize = 15)
pyplot.xlabel('City Name',fontsize = 15)
pyplot.show()
```

```python
#for tweets mentioning trump
figure(figsize=(20, 15))
pyplot.ylim(0, 5)
pyplot.boxplot(list(ratio_record_trump.values())[:30],labels=list(ratio_record_trump.keys())[:30], notch = False)
pyplot.title('Boxplot for ratio of fake news against true news for different cities (trump)',fontsize = 20)
pyplot.ylabel('Ratio of fake to true news',fontsize = 15)
pyplot.xlabel('City Name',fontsize = 15)
pyplot.show()
```

Different cities exert very different fake to true news ratio.  For majority days of the 21 day period, most cities seem to produce more true news than fake news while some cities seem to produce more fake news than true news.
This clearly shows cities can be used as an attribute to classify fake news and true news. An aside is that if we dig deep into this, we can probably discover which city supports which candidate.

### Tweet Creation Time

```python
#creation time

def get_hour(created_at):
    
    '''The hour when the tweet was created'''
    return int(created_at[11:13])

get_hour_udf = udf(lambda x: get_hour(x), IntegerType())
```

```python
def hour_useful(folder):
    '''Return the median of fake to true news ratio for all 24 hours for the 21 days period.
    input is a directory, output are two dictionaries, keys are cities, values are lists
    '''    
    
    ratio_record = {}

    for i in range(21):  
        hour_dict = {}
        data_from_file = spark.read.json(folder+files[i])
        new_dataframe = data_from_file.withColumn('hour',get_hour_udf(data_from_file.created_at)).filter('fact_check==1 or fact_check == -1').groupby('hour','fact_check')
        records = new_dataframe.agg(func.count(func.lit(1)).alias("Num Of Records"))

        true_record = records.filter('fact_check==1').collect()
        false_record = records.filter('fact_check==-1').collect()

        for j in true_record:
            temp_hour = j[0]
            if temp_hour not in hour_dict:
                hour_dict[temp_hour] = j[2]
            
        for k in false_record:
            temp_hour = k[0]
            if temp_hour in hour_dict:
                if temp_hour in ratio_record:
                    ratio_record[temp_hour].append(k[2]/hour_dict[temp_hour])                
                else:
                    ratio_record[temp_hour] = [k[2]/hour_dict[temp_hour]]
        
    return  ratio_record
    
```

```python
ratio_record_hillary = hour_useful(hillary_folder)
ratio_record_trump = hour_useful(trump_folder)
```

```python
#plotted the results for tweets mentioning hillary
figure(figsize=(20, 15))
pyplot.boxplot(list(ratio_record_hillary.values()),labels=list(ratio_record_hillary.keys()), notch = False)
pyplot.title('Boxplot for ratio of fake news against true news for different hours (hillary)',fontsize = 20)
pyplot.ylabel('Ratio of fake to true news',fontsize = 15)
pyplot.xlabel('Hour',fontsize = 15)
pyplot.show()
```

```python
#plotted the results for tweets mentioning trump
figure(figsize=(20, 15))
pyplot.boxplot(list(ratio_record_trump.values()),labels=list(ratio_record_trump.keys()), notch = False)
pyplot.title('Boxplot for ratio of fake news against true news for different hours (trump)',fontsize = 20)
pyplot.ylabel('Ratio of fake to true news',fontsize = 15)
pyplot.xlabel('Hour',fontsize = 15)
pyplot.show()
```

From the plots, it seems hour cannot be used as a feature to distinguish fake news from true news. The distribution of ratio of fake news to true news appear to be very even throughout the 21 days period.

## User level features

### User Verification

The first feature I will explore is user verification. Users can be offically verified by Twitter if it is detemrined as an account of public interest, like musicians and singers.

I will calculate the ratio of fake to true news for the group of verified users and unverified users.

```python
#verified, unverified user for fake news
#for hillary and trump

def verified_unverified(folder):
    
    '''input a folder, read all the files inside and return the ratio of fake to true news 
    for verified/unverified users across the period
    '''
    
    ratio_for_verified = [0]*21
    ratio_for_unverified = [0]*21
    

    for i in range(21):
        data_from_file = spark.read.json(folder+files[i])
        number_of_fake_verified = data_from_file.filter('fact_check==-1').filter('user_verified==True').count()
        number_of_true_verified = data_from_file.filter('fact_check==1').filter('user_verified==True').count()
        number_of_fake_unverified = data_from_file.filter('fact_check==-1').filter('user_verified==False').count()
        number_of_true_unverified = data_from_file.filter('fact_check==1').filter('user_verified==False').count()
        ratio_for_verified[i] = number_of_fake_verified/number_of_true_verified
        ratio_for_unverified[i] = number_of_fake_unverified/number_of_true_unverified
        
    return ratio_for_verified,ratio_for_unverified
```

```python
#number of true/fake news url passed

ratio_for_verified_hillary,ratio_for_unverified_hillary = verified_unverified(hillary_folder)
ratio_for_verified_trump,ratio_for_unverified_trump = verified_unverified(trump_folder)
```

```python
figure(figsize=(20, 15))
pyplot.subplot(2, 1, 1)
y_axis = ratio_for_verified_hillary
pyplot.plot( x_axis, y_axis, '-', label = 'verified' )

y_axis = ratio_for_unverified_hillary
pyplot.plot( x_axis, y_axis, '-',label = 'unverified' )

ax = pyplot.gcf().axes[0] 
ax.xaxis.set_major_formatter(formatter)
pyplot.gcf().autofmt_xdate(rotation=25)
pyplot.legend(loc='best')
pyplot.title('Ratio of Fake News to True News for verfied/unverified user (Hillary)')

pyplot.subplot(2, 1, 2)
y_axis = ratio_for_verified_trump
pyplot.plot( x_axis, y_axis, '-', label = 'verified' )

y_axis = ratio_for_unverified_trump
pyplot.plot( x_axis, y_axis, '-',label = 'unverified' )

ax = pyplot.gcf().axes[0] 
ax.xaxis.set_major_formatter(formatter)
pyplot.gcf().autofmt_xdate(rotation=25)
pyplot.legend(loc='best')
pyplot.title('Ratio of Fake News to True News for verfied/unverified user (Trump)')


pyplot.show()
```

For tweets mentioning Hillary ot Trump, it is clear those produced by unverified users have a far higher fake to true news ratio than those produced by verified users. The ratio of fake to true news for tweets mentioning Hillary or Trump is about 0.01 for verified users. And the ratio of fake to true news mentioning Hillary for unverified users was between 0.2 and 0.4. The ratio of fake to true news for tweets mentioning Hillary or Trump is about 0.01 for verified users. And the ratio of fake to true news mentioning Hillary for unverified users was between 0.07 and 0.2.

### Other user features
Other user features I considered were user_followers_count, user_friends_counnt, user_favourites_count, user_statuses_count. 
Here I built a logistic regression model with lasso penalty to conduct feature selection.

```python
#change fact_check value to 0 (fake) and 1 (true)
def update_fact_check(fact_check):
    return (fact_check+1)/2
        
get_year_udf = udf(lambda x: update_fact_check(x), IntegerType())
```

```python
#First put data into correct format

data_from_file = spark.read.json("gs://good-bucket/st446_project/hillary/2016-10-13.json")
other_user_attribute = ['user_followers_count','user_friends_count','user_favourites_count','user_statuses_count']
data_from_file = data_from_file.filter('fact_check==1 or fact_check==-1').withColumn("followers_count", data_from_file["user_followers_count"].cast(IntegerType())).withColumn("friends_count", data_from_file["user_friends_count"].cast(IntegerType())).withColumn("favourites_count", data_from_file["user_favourites_count"].cast(IntegerType())).withColumn("statuses_count", data_from_file["user_statuses_count"].cast(IntegerType())).select(*other_user_attribute,'fact_check')
data_from_file = data_from_file.withColumn('fact_check2',update_fact_check(data_from_file.fact_check)).select(*other_user_attribute,'fact_check2')
```

```python
data_from_file.printSchema()
```

```python
import pyspark.ml.feature as ft
import pyspark.ml.classification as cl
from pyspark.ml import Pipeline
import pyspark.ml.evaluation as ev
```

```python
#Create Transformer
featuresCreator = ft.VectorAssembler(
    inputCols= data_from_file.columns[:-1],
    outputCol='features'
)
#Create an estimator
logistic = cl.LogisticRegression(
    maxIter=10, 
    regParam=0.1, 
    labelCol='fact_check2',  elasticNetParam = 0) 
#Create a pipeline
pipeline = Pipeline(stages=[
        featuresCreator, 
        logistic
    ])
model = pipeline.fit(data_from_file)
test_model = model.transform(data_from_file)

#Evaluate the performance
evaluator = ev.BinaryClassificationEvaluator(
    rawPredictionCol='probability', 
    labelCol='fact_check2')

print('areaUnderROC: ', evaluator.evaluate(test_model, 
     {evaluator.metricName: 'areaUnderROC'}))
print('areaUnderPR: ', evaluator.evaluate(test_model, {evaluator.metricName: 'areaUnderPR'}))
```

### Conclusion

#### Tweet level features:
* The city where the tweet was sent is significant in classifying tweets. Different cities exert very different fake to true news ratio.  For majority days of the 21 day period, most cities seem to produce more true news than fake news while some cities seem to produce more fake news than true news. 
* Hour when the tweeted was created cannot be used as a feature to distinguish fake news from true news. The distribution of ratio of fake news to true news for any hour appear to be very similiar throughout the 21 days period.

#### User level features:
* User verification is an important feature in classifying news. For tweets mentioning Hillary ot Trump, it is clear those produced by unverified users have a far higher fake to true news ratio than those produced by verified users. 

* We can use year of creation (whether created just before a major political event or not) as a factor to classify political fake news. The number of fake news published by accounts created in 2016 is far larger (about 3 times) than that created by account in any other year. 

* Friends_count is an important feature in classifying news. Also, any pair of favourite_count, follower_count,statues_count would also help (any third one is redudant). For the logistic regression model that uses these 4 features alone, we can an areaUnderROC of 0.57 and an area under PR of 0.81.
